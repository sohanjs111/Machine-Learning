# Decision Tree Classification

## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

## Importing the dataset

```python
dataset = pd.read_csv('https://raw.githubusercontent.com/sohanjs111/Machine-Learning/main/Database/Social_Network_Ads.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
dataset.head()
```

## Splitting the dataset into the Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

## Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
s = StandardScaler()
s.fit(X_train)

X_train = s.transform(X_train)
X_test  = s.transform(X_test)
```

## Training the Decision Tree Classification model on the Training set

```python
from sklearn.tree import DecisionTreeClassifier
classr = DecisionTreeClassifier(criterion = 'entropy', random_state=0)
classr.fit(X_train,y_train)
```

## Predicting a new result

```python
y_pr = classr.predict(X_train)
y_pr
```

## Predicting the Test set results

```python
y_pred = classr.predict(X_test)
y_pred
```

## Making the Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

## Visualising the Training set results

```python
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline


cls = tree.DecisionTreeClassifier(random_state=0)


cls = cls.fit(X_train, y_train)
plt.figure(figsize=(15,10))
tree.plot_tree(cls,filled=True)
```

## Visualising the Test set results

```python
cls = cls.fit(X_test, y_test)
plt.figure(figsize=(15,10))
tree.plot_tree(cls,filled=True)
```
