```python
import pandas as pd
import math
import numpy

from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor 
from sklearn.preprocessing import MinMaxScaler, LabelEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv("50_Startups.csv")
df.head()
```

```python
#Using OneHotEncoder to change string values to integer
from sklearn.preprocessing import OneHotEncoder
# Converting type of columns to category
df['State']=df['State'].astype('category')

#Assigning numerical values and storing it in another columns
df['newstates']=df['State'].cat.codes

#Create an instance of One-hot-encoder
enc=OneHotEncoder()
enc_data=pd.DataFrame(enc.fit_transform(df[['newstates']]).toarray())
#print(enc_data)

#Merge with the dataset
df.join(enc_data)
df.head()
```

```python
df['profit'] = df['Profit']
df.drop(columns=['Profit' , 'State' ], axis=1,inplace=True)
# df.drop(columns=[ 'State' , 'Administration',  'Marketing Spend' ], axis=1,inplace=True)
df.head()
```

```python
# X = df.iloc[:,0:-1].values
# y = df.iloc[:,-1].values

X = df.drop(['profit'], axis=1)
y = df['profit'].values
print(type(X))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=2)
print(X_train)
print(y_train)
```

```python
model = RandomForestRegressor(n_estimators=30, max_depth=2)
model.fit(X_train, y_train)
```

```python
ypred = model.predict(X_test)
print(ypred)
```

```python
print("Diffrence between Actual and Predicted Profit is : \n")
for i in range(len(ypred)):
    print(f"{y_test[i]} - {ypred[i]} = {math.floor(y_test[i] - ypred[i])}")
```

```python
from sklearn.metrics import r2_score

print(f'Accuracy: { r2_score(y_test, ypred) }')
```
