# Simple Linear Regression

## Import the libraries

```python
import numpy as np
import pandas as pd
```

## Import the dataset

```python
df= pd.read_csv("income.data.csv")
df.head()
```

```python
df.count()
```

```python

from sklearn.model_selection import train_test_split

x=df.iloc[:,1:-1].values
y=df.iloc[:,-1].values

x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2,random_state=50 )
```

```python
print(x.size)
print(y.size)
print(x.shape,x_train.shape,x_test.shape)
print(y.shape,y_train.shape,y_test.shape)
```

# Visualizing the data

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(df.income, df.happiness,marker='x', c='r')
```

```python
import seaborn as sns
sns.pairplot(data=df,hue= "income")
```

```python
#correlation between heatmap helps ous to find interrelation between each and every feature

sns.heatmap(data=df.corr())
```

```python
df.isnull().sum()
```

## Training the Simple Linear Regression model on the Training set

```python
from sklearn.linear_model import LinearRegression

model=LinearRegression()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```

## Predicting the Test set results

```python
print(x_train.size)
print(y_train.size)
print(x_test.size)
print(y_test.size)
```

```python
 y_pred = model.predict(x_test)
 y_pred
```


# Comparing the Test Set with Predicted Values

```python
data = pd.DataFrame({'Real Values':y_test, 'Predicted Values':y_pred})
data

#We can see that the predicted salaries are very close to the real salary values and it 
#can be concluded that the model has been well trained.
```

## Visualising the Training set results

```python
train_pred = model.predict(x_train)
plt.scatter(x_train, y_train)                #Plot the datapoints
plt.plot(x_train, train_pred,color='black')         #plots the model prediction
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.title('Training set')
```

## Visualising the Test set results

```python
plt.scatter(x_test, y_test)
plt.plot(x_train, train_pred, color='Black')
plt.xlabel('Income')
plt.ylabel('Happiness')
plt.title('Testing set')
```
